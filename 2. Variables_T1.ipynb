```python
# Integer Variable
age = 25 
print(age)
```

```python
# String Variable
name = 'Alice'
print(name)
```

```python
# Float Variable  
price = 19.99
print(price)
```

```python
# Boolean Variable
is_active = True 
print(is_active)
```

```python
x = 10
# printing the value of 'x'
print(x) # output: 10 
```

```python
# Assigning values to variables

a = 5
b = 3
# adding two variables and storing the result in 'result'
result = a + b
# printing the result
print(result)
```

```python
# initial value
score = 50
print(score) 

# changing the value of 'score'
score = 100
print(score)
```

```python
# assigning values to variables 
first_name = "John"
last_name = "Doe"

# concatenating strings and storing in a variable
full_name = first_name + " " + last_name
print(full_name) 
```

```python
# Using Variables in a calculation

# assigning values to variables
length = 10
breadth = 5

# calculating the area of rectangle 
area = length * breadth
print(area)
```
